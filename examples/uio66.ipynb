```python
from mofbuilder import MetalOrganicFrameworkBuilder as MofBuilder
import veloxchem as vlx

smi_str = {
    "ndi":
    "OC(C(C=C1)=CC(O)=C1N(C2=O)C(C3=CC=C4C5=C3C2=CC=C5C(N(C6=C(O)C=C(C(O)=O)C=C6)C4=O)=O)=O)=O"
}
smi_str["bdc"] = "O=C([O-])C(C=C1)=CC=C1C([O-])=O"
smi_str["135btc"] = "C1=C(C=C(C=C1C(=O)[O-])C(=O)[O-])C(=O)[O-]"
```

```python
mof = MofBuilder(mof_family="UiO-66")
mol = vlx.Molecule.read_smiles(smi_str["bdc"])

mof.linker_molecule = mol
mof.node_metal = "Zr"
mof.supercell = (1, 1, 1)
original_mof = mof.build()
```

```python
original_mof.md_prepare()
```

```python
original_mof.write(format=["gro"], filename="uioc66_original.gro")
```

```python
original_mof.md_driver.run_pipeline(
    steps=['em', 'nvt', 'npt'],
    nvt_time=10,
    npt_time=0,
    output_prefix="MOF_MD",
    record_interval=1,  # record every 1 ps
    whole_traj=False  # output single continuous PDB
)
```
