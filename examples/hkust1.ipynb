```python
from mofbuilder import MetalOrganicFrameworkBuilder as MofBuilder
import veloxchem as vlx

smi_str = {
    "ndi":
    "OC(C(C=C1)=CC(O)=C1N(C2=O)C(C3=CC=C4C5=C3C2=CC=C5C(N(C6=C(O)C=C(C(O)=O)C=C6)C4=O)=O)=O)=O"
}
smi_str["bdc"] = "O=C([O-])C(C=C1)=CC=C1C([O-])=O"
smi_str["135btc"] = "C1=C(C=C(C=C1C(=O)[O-])C(=O)[O-])C(=O)[O-]"
```

```python
mof = MofBuilder(mof_family="HKUST-1")
```

```python
mof.list_available_metals()
```

```python
mof = MofBuilder(mof_family="HKUST-1")
mol = vlx.Molecule.read_smiles(smi_str["135btc"])
mof.linker_molecule = mol
#mof.save_files = True
#mof.linker_xyzfile = "tests/testdata/testtrilinker.xyz"
mof.load_optimized_rotations = "hkust_rot.h5"
mof.skip_rotation_optimization = True
#mof.rotation_filename = "hkust_rot"
#mof._debug = True
#mof.termination = False
mol.show()

mof.node_metal = "Cu"
mof.supercell = (1, 1, 1)
original_mof = mof.build()
```

```python
original_mof.write(format=["gro", "xyz"], filename="hku")
```

```python
original_mof.show()
```

```python
original_mof.resp_charges = False
original_mof.md_prepare()
original_mof.md_driver.run_pipeline(
    steps=['em', 'nvt', 'npt'],
    nvt_time=10,
    npt_time=0,
    output_prefix="hkust1_MD",
    record_interval=1,  # record every 1 ps
    whole_traj=False  # output single continuous PDB
)
```
