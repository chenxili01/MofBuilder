```python
from mofbuilder import MetalOrganicFrameworkBuilder as MofBuilder
import veloxchem as vlx
mof = MofBuilder(mof_family="MIL-101")
#mof.data_path = '/home/vlxuser/envs/MofBuilder/database' # path to the database should be given
mol = vlx.Molecule.read_smiles("O=C([O-])C(C=C1)=C(N)C=C1C([O-])=O")
mof.target_directory = 'test'
mof.linker_molecule = mol
mof.rotation_filename='mil101.h5'
mof.load_optimized_rotations='mil101.h5'
mof.skip_rotation_optimization= True
mof.node_metal = "Cr"
mof.dummy_atom_node = True

mof.supercell = [1, 1, 1]
mof.cleaved_eG=[[0.1,0.3],[0.1,0.3],[0.1,0.3]]
original_mof = mof.build()
```
